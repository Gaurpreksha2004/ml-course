# The $\mathbf{numpy.linspace()}$ Function 

The $\mathbf{numpy.linspace()}$ function in Python's NumPy library is used to **create an array of evenly spaced numbers over a specified interval.**

---

## How it Works

The function takes at least three main arguments:

* **`start`**: The starting value of the sequence.
* **`stop`**: The end value of the sequence.
* **`num`**: (Optional, default is 50) The number of samples (data points) to generate.

It returns a NumPy array where the points are linearly spaced between the `start` and `stop` values. The spacing between any two consecutive values is **constant**.

---

## Key Characteristics

* **Inclusive Stop:** By default, the `stop` value **is included** in the array.
* **Constant Step:** It calculates the constant step size required to divide the interval $[start, stop]$ into $\mathbf{(num-1)}$ equal parts.

### Example

```python
import numpy as np

# Create 5 points from 0 to 10
array_1 = np.linspace(0, 10, 5)

print(array_1)
# Output: [ 0.   2.5  5.   7.5 10. ]

#  Explanation of `.keys()` in `ax.bar(nut_butter_prices.keys(), ...)`

---

##  What is `nut_butter_prices.keys()`?

The object `nut_butter_prices` is a standard **Python Dictionary**. Dictionaries store data as **key-value pairs**.

| **Code** | `nut_butter_prices = {"Almond Butter": 12.00, "Peanut Butter": 6.50}` |
| :--- | :--- |
| **Keys (Categories)** | `"Almond Butter"`, `"Peanut Butter"` |
| **Values (Data Points)** | `12.00`, `6.50` |

The `.keys()` method returns a **view object** containing all the keys of the dictionary.

---

##  Role in the `ax.bar()` Function

The `ax.bar()` function from Matplotlib is used to draw a bar chart and typically takes two main arguments: `x` and `height`.

| Function Parameter | Argument Provided | Purpose in the Chart |
| :--- | :--- | :--- |
| **`x`** (The X-coordinates) | `nut_butter_prices.keys()` | These are the **labels** for the bars, which appear along the **x-axis** (e.g., the names of the nut butters). |
| **`height`** (The Y-coordinates) | `nut_butter_prices.values()` | These are the **lengths/magnitudes** of the bars, determining their height on the **y-axis** (e.g., the prices). |

### Summary

In this line of code, **`.keys()` is responsible for providing the categorical labels for the horizontal (X) axis.**

###  Difference between `np.random.random()` and `np.random.randn()`

####  `np.random.random()`
- **Definition:** Generates random float numbers between **0.0 and 1.0**.  
- **Distribution Type:** **Uniform distribution**  All values in [0, 1) are equally likely.  
- **Syntax:**
  ```python
  np.random.random(size)
  ```
- **Example:**
  ```python
  np.random.random(3)
  # Output: array([0.45, 0.87, 0.12])
  ```
- **Range:** [0.0, 1.0)

---

####  `np.random.randn()`
- **Definition:** Generates random numbers from a **standard normal distribution** (mean = 0, std = 1).  
- **Distribution Type:** **Normal (Gaussian) distribution**  Bell curve centered at 0.  
- **Syntax:**
  ```python
  np.random.randn(size)
  ```
- **Example:**
  ```python
  np.random.randn(3)
  # Output: array([-0.54,  0.32,  1.15])
  ```
- **Range:** Unbounded (can be any real number, but most values lie between -3 and 3)

---

####  Summary Table

| Function | Distribution Type | Range | Mean | Example Output |
|-----------|------------------|--------|------|----------------|
| `np.random.random()` | Uniform | [0, 1) | 0.5 | [0.12, 0.89, 0.33] |
| `np.random.randn()` | Normal (Gaussian) | (-, ) | 0 | [-0.45, 0.67, -1.23] |

---

####  Tip:
To generate values from a **normal distribution with a custom mean and standard deviation**:
```python
mu, sigma = 5, 2
data = mu + sigma * np.random.randn(1000)
```
This gives numbers with **mean  5** and **standard deviation  2**.

#  `cumsum()` Function in Pandas

The **`cumsum()`** function in **Pandas** stands for **cumulative sum**.  
It returns the **running total** (cumulative sum) of the values along a specified axis in a **Series** or **DataFrame**.

---

###  Syntax
```python
DataFrame.cumsum(axis=None, skipna=True)

Purpose: Adds each element to all previous elements.
Result: Each value represents the sum of all elements up to that point.

 Example 1: On a Series
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s.cumsum())

Output:
0     10
1     30
2     60
3    100
dtype: int64

Explanation:

10  10

10 + 20  30

10 + 20 + 30  60

10 + 20 + 30 + 40  100

 Example 2: On a DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [10, 20, 30]
})
print(df.cumsum())

Output:
   A   B
0  1  10
1  3  30
2  6  60
