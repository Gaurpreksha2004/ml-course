## What is a Crosstab?

A crosstab (cross-tabulation) is just a table that shows how two columns are related.

It counts how many times each combination of values appears.

Think of it like making a mini frequency table for two variables at the same time.

___________________________________________________________________________________________

ðŸ”¹ Example in real life

Suppose you survey some people about their Gender and whether they Own a Car:
| Person | Gender | Car |
| ------ | ------ | --- |
| 1      | M      | Yes |
| 2      | F      | No  |
| 3      | M      | Yes |
| 4      | F      | Yes |
| 5      | M      | No  |

Now, a crosstab of Gender vs Car looks like this:

| Gender | No | Yes |
| ------ | -- | --- |
| F      | 1  | 1   |
| M      | 1  | 2   |

ðŸ”¹ How to read this table?

For F (Female) â†’ 1 said No, 1 said Yes.

For M (Male) â†’ 1 said No, 2 said Yes.

So, the table quickly shows the relationship between Gender and Car Ownership.

```python
# Create a crosstab of the Make and Doors columns

pd.crosstab(car_sales["Make"], car_sales["Doors"], normalize = 'index')
```

## What does normalize='index' do?

Normally, pd.crosstab() gives counts (how many times each combination appears).
But sometimes we donâ€™t just want counts â€” we want percentages (proportions).

normalize='index' â†’ converts the counts row-wise into proportions (i.e., divide each row by its row total).

So, instead of raw counts, you get fractions or percentages that add up to 1 for each row.

```python
# Create a column which represents the price of a car per kilometer
# Then view the DataFrame

# Force Price column to string first, then clean
car_sales_missing['Price'] = (
    car_sales_missing['Price']
    .astype(str)                             # make sure all values are strings
    .str.replace('$', '', regex=False)       # remove dollar sign
    .str.replace(',', '', regex=False)       # remove commas
)

# Convert cleaned values to integer
car_sales_missing['Price'] = car_sales_missing['Price'].astype(int)

car_sales_missing['Price (per KM)'] = car_sales_missing['Price'] / car_sales_missing['Odometer'] 

car_sales_missing
```

```python
# Remove the last column you added using .drop()
car_sales_missing = car_sales_missing.drop('Price (per KM)', axis = 1)
car_sales_missing
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the the shuffled DataFrame to a new variable

shuffled_df = car_sales_missing.sample(frac = 1, random_state = 42)
shuffled_df.head()
```

ðŸ”¹ Explanation

frac=1 â†’ means return 100% of the rows but in a random order.

random_state=42 â†’ makes the shuffle reproducible (youâ€™ll get the same shuffled order each time you run it). 
                  If you leave it out, the shuffle will be different every run.

Notice how the index numbers get moved around. The [`sample()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html) function is a great way to get random samples from your DataFrame. It's also another great way to shuffle the rows by setting `frac=1`.

```python
# Reset the indexes of the shuffled DataFrame
shuffled_df.reset_index(drop = True, inplace = True)
shuffled_df
```
